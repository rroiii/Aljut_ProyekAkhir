Kolom yang memiliki missing value adalah Gross, Meta-score, dan Certificate.
akan dilakukan beberapa operasi untuk mengolah data yang hilang (bernilai NaN / NULL):
1. Baris yang data Grossnya bernilai NULL akan di-drop karena Gross adalah nilai yang akan dianalisa
2. Baris yang data Certificatenya bernilai NULL akan di-drop karena Certificate akan digunakan sebagai dependensi analisis
3. Baris yang data Meta-scorenya bernilai NULL akan diganti menggunakan nilai Meta-score rata-rata

Menghapus baris yang tidak memiliki nilai Gross atau Gross==NULL (akan menghapus 169 data sehingga data setelah dihapus = 831 data)

```python
# simply drop whole row with NaN in "price" column
ext.dropna(subset=["Gross"], axis=0, inplace=True)
# reset index, because we droped two rows
ext.reset_index(drop=True, inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
ext.dropna(subset=["Certificate"], axis=0, inplace=True)
# reset index, because we droped two rows
ext.reset_index(drop=True, inplace=True)
```

Menghitung nilai rata-rata "Meta_score"

```python
avg_meta_score = ext['Meta_score'].astype("float").mean(axis=0)
print("Rata-rata meta-score adalah :", avg_meta_score)
```

Mereplace data yang memiliki Meta_score==NULL dengan nilai rata-rata Meta_score film yang memiliki Certificate dan Gross

```python
ext["Meta_score"].replace(np.nan, avg_meta_score, inplace=True)
```
